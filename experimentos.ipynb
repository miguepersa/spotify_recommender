```python
from recommendator import recommendation, get_song_data, final_df, ind, km
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
song1 = get_song_data("167UNsZbooXaxiFYaLGYBr")
recommendations1 = recommendation(song1, 5)
recommendations1_data = [final_df[final_df['id'] == r[1]] for r in recommendations1]
df = pd.DataFrame(recommendations1_data[0][ind])
for i in range(1,len(recommendations1_data)):
    for _,row in recommendations1_data[i][ind].iterrows():
        df.loc[len(df.index)] = row

df.loc[len(df.index)] = song1
sns.pairplot(df, palette=sns.color_palette("dark"))
plt.savefig("./plots/song1.png")
```
