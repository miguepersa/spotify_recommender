# Elbow Method

```python
from K_means import Kmeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inertias = []
ind = ["acousticness","danceability","energy","instrumentalness","key","liveness","loudness","speechiness","tempo","valence"]
music_data = pd.read_csv("music_data_popular.csv")

for i in range(5,100,5):
    km = Kmeans(music_data[ind], i)
    km.train(1000)
    inertias.append(km.inertia())
```

```python
plt.plot(range(5,25,5), inertias)
plt.show()
plt.savefig("./plots/elbow.png")
```
