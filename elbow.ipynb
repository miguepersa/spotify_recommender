# Elbow Method

```python
from K_means import Kmeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inertias = []
ind = ['danceability','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo', 'time_signature']
music_data = pd.read_csv('genres_v2.csv', low_memory=False)

for i in range(5,25,5):
    km = Kmeans(music_data[ind], i)
    km.train(100)
    inertias.append(km.inertia())
```

```python
plt.plot(range(5,25,5), inertias)
plt.show()
plt.savefig("./plots/elbow.png")
```
